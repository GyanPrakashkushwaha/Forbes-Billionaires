```python
category_counts = df['business_category'].value_counts()
rare_categories = category_counts[category_counts < 10].index.tolist()
print(rare_categories)
```

```python
def make_attribute(text):
    if text in rare_categories:
        return 'others'
    else:
        return text
```

```python
df['business_category'] = df['business_category'].apply(func=make_attribute)
```
