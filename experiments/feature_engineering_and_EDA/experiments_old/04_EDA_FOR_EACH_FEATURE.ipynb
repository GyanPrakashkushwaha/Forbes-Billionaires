```python
def random_values(feature_name:str,total=6):
    for i in range(0,total):
        index = np.random.randint(df.shape[0])
        print(index,':',df[feature_name].iloc[index])
```

```python
def all_about_col(col_name:str):
    print(f'missing values in {col_name} : {df[col_name].isnull().sum()}')
    print()
    print(f'value counts : {df[col_name].value_counts()}')
    
```
